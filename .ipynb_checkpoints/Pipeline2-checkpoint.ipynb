```python
import pandas as pd
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
db = client["spotify_db"]

artists_col = db.spotify_artists
tracks_col = db.spotify_artist_tracks
combined_col = db.spotify_artist_top10
```

```python
artists_df = pd.read_json("Spotify/spotify_artist_info_20251201.json")

artists_col.delete_many({})
artists_col.insert_many(artists_df.to_dict(orient="records"))
```

```python
tracks_df = pd.read_json("Spotify/spotify_artist_tracks_20251201.json")
tracks_col.delete_many({})
tracks_col.insert_many(tracks_df.to_dict(orient="records"))
```

```python
import json
with open("filtered_artist_ids.json", "r") as f:
    filtered_ids = json.load(f)
```

```python
pipeline = [

    # 1. Filter only selected artists (FASTEST to do first)
    {
        "$match": {
            "artist_id": {"$in": filtered_ids}
        }
    },

    # 2. Join with the top tracks collection
    {
        "$lookup": {
            "from": "spotify_artist_tracks",
            "localField": "artist_id",
            "foreignField": "artist_id",
            "as": "tracks"
        }
    },

    # 3. Sort tracks by popularity (DESC) and keep only the top 10
    {
        "$addFields": {
            "tracks": {
                "$slice": [
                    {
                        "$sortArray": {
                            "input": "$tracks",
                            "sortBy": { "popularity": -1 }
                        }
                    },
                    10
                ]
            }
        }
    },

    # 4. Final shape of output
    {
        "$project": {
            "_id": 0,             # remove MongoDB _id
            "artist_id": 1,        # PRIMARY KEY
            "name": 1,
            "genres": 1,
            "followers": 1,
            "popularity": 1,
            "tracks.track_id": 1,
            "tracks.track_name": 1,
            "tracks.popularity": 1,
            "tracks.album_name": 1,
            "tracks.release_date": 1
        }
    }
]
```

```python
combined = list(artists_col.aggregate(pipeline))

combined_col.delete_many({})
combined_col.insert_many(combined)

print("Combined artist+top10 tracks stored successfully.")
```

```python
import pandas as pd

# Run the aggregation pipeline

rows = []

# Flatten structure for CSV
for doc in combined:
    artist_id = doc["artist_id"]
    name = doc.get("name")
    genres = ", ".join(doc.get("genres", []))
    followers = doc.get("followers")
    popularity = doc.get("popularity")

    for t in doc["tracks"]:
        rows.append({
            "artist_id": artist_id,
            "artist_name": name,
            "genres": genres,
            "followers": followers,
            "artist_popularity": popularity,
            "track_id": t.get("track_id"),
            "track_name": t.get("track_name"),
            "track_popularity": t.get("popularity"),
            "album_name": t.get("album_name"),
            "release_date": t.get("release_date")
        })

# Convert to DataFrame
df = pd.DataFrame(rows)

# Export to CSV
# df.to_csv("spotify_artists_with_top10_tracks.csv", index=False)

df.head()
```
